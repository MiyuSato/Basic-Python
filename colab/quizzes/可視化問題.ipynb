## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

問1.png

```python
df = pd.read_csv('matplotlib_quiz/hist_data.csv')

fig = plt.figure(figsize = (5,3))
plt.title('aデータとbデータの温度に関する度数分布', fontsize = 10)
plt.xlabel('温度(℃)', fontsize = 8)            
plt.ylabel('度数', fontsize = 8) 

plt.hist(df['aデータ'], bins = 30, alpha = 0.5, color = '#ffaf00', label = 'aデータ')
plt.hist(df['bデータ'], bins = 30, alpha = 0.4, color = '#00dcd4', label = 'bデータ')

plt.legend(loc = 'upper right', fontsize = 8)
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

問2.png

```python
import seaborn as sns

df = pd.read_csv('matplotlib_quiz/scatter_data.csv')

fig = plt.figure(figsize = (5,3))
plt.title('クラス０〜２のlengthとwidthに関する散布図', fontsize = 10)
plt.xlabel('length (cm)', fontsize = 8)            
plt.ylabel('width (cm)', fontsize = 8) 

sns.scatterplot(df, x='length (cm)', y='width (cm)', hue='class',palette="bright")

plt.legend(labels=["class0","class1","class2"], loc = 'upper right', fontsize = 8)
plt.grid()
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

問3(1).png

問3(2).png

```python
df = pd.read_csv('matplotlib_quiz/line_data.csv')

fig = plt.figure(figsize = (5.5,3.5))

plt.title('期間ごとの件数の推移', fontsize = 11)
plt.xlabel('期間', fontsize = 9)            
plt.ylabel('件数', fontsize = 9) 

plt.plot('期間', '件数', data = df, marker='o')

plt.tick_params(axis = 'x', rotation = 30, labelsize = 9)
plt.grid()
```

```python
fig = plt.figure(figsize = (5.5,3.5))

plt.title('期間ごとの費用の推移', fontsize = 11)
plt.xlabel('期間', fontsize = 9)            
plt.ylabel('費用(円)', fontsize = 9) 

plt.bar('期間', '費用', data = df)

plt.tick_params(axis = 'x', rotation = 30, labelsize = 9)
plt.grid()
```
